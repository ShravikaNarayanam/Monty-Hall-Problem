```python
import random 
import pandas as pd

#counter variables for win_stay and win_switch
win_stay = 0
win_switch = 0

#Function to select a random door 
def random_door():
    random_door.random_choice = random.randint(1,3)
    return random_door.random_choice

#Function to select car door
def select_car_door():
    select_car_door.car_door = random_door()
    return select_car_door.car_door

#Function to select user door
def user_select_door():
    user_select_door.user_choice_door = random_door()
    return user_select_door.user_choice_door

#Function to select which door monty opens
def opened_door():
    #door has the value generated by random_door(random door number)
    door = random_door()
    while door == user_select_door.user_choice_door or door == select_car_door.car_door:
        door = random_door()
    #Door opened by montey which satisfies the above condition
    opened_door.open_door = door
    return opened_door.open_door

#Function deciding win based on stay or switch
def stay_or_switch():
    global win_stay
    global win_switch
    global user_second_choice
    
    stay_or_switch.win_if_switch = ""
    stay_or_switch.win_if_stay = ""
    #condition for stay
    if user_select_door.user_choice_door == select_car_door.car_door:
        win_stay += 1
        stay_or_switch.win_if_stay = "yes"
    else:
        #condition for switch
        user_second_choice = 6 - user_select_door.user_choice_door - opened_door.open_door
        if user_second_choice == select_car_door.car_door:
            win_switch += 1
            stay_or_switch.win_if_switch = "yes"
    return win_stay, win_switch,user_second_choice

#Function to play the game
def play_game():
    select_car_door()
    user_select_door()
    opened_door()
    stay_or_switch()

#Print column names
print("Index CarDoor UserDoor OpenDoor SwitchDoor   Switch?   Stay?\n ")
#Range of the simulations = 100
for i in range(100):
    play_game()
    print(f'{i:2}{select_car_door.car_door:10}{user_select_door.user_choice_door:10}{opened_door.open_door:10}{user_second_choice:10}   {stay_or_switch.win_if_switch:10}  {stay_or_switch.win_if_stay:10}')

print("win count when player stays with first choice:",win_stay)
print("win count when player switch to a second choice:",win_switch)

#Calculating win ratio
win_ratio = (win_switch / win_stay)
print("Win ratio of switching over staying:{:.2}".format(win_ratio))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#per_stay = win_stay/(win_stay + lost_stay)
#per_switch = win_switch/(win_switch + lost_switch)
per_stay = win_stay/(win_stay + win_switch)
per_switch = win_switch/(win_stay + win_switch)
xaxis = ['Stay', 'Switch']
yaxis = [win_stay, win_switch]
y_topper = [per_stay,per_switch]

sns.set_style('whitegrid')  # white backround with gray grid lines
axes = sns.barplot(xaxis, yaxis, palette='bright')
axes.set_title('Stay vs Switch win percentages after 100 simulation')
axes.set(ylabel = "Simulations")

# Scale the y-axis by 10% to make room for text above the bars.
axes.set_ylim(top=1.10*max(yaxis))

# Display the count above each patch (bar).
for bar, count in zip(axes.patches, y_topper):
    text_x = bar.get_x() + bar.get_width()/2  
    text_y = bar.get_height() 
    text   = f'{count:.2%}'
    axes.text(text_x, text_y, text, 
              fontsize=11, ha='center', va='bottom')

plt.show()  # display the chart 
```
